```python
import pickle
import pandas as pd
import numpy as np
```

```python
for user in range(1, 25):
    with open(f'data/motion_sense/{user}_train.pickle', 'rb') as train_file:
        train = pickle.load(train_file)
        
    with open(f'data/motion_sense/{user}_test.pickle', 'rb') as test_file:
        test = pickle.load(test_file)
        
    y_train = train['activity'].values
    train.drop('activity', axis=1, inplace=True)
    x_train = train.values
    
    y_test = test['activity'].values
    test.drop('activity', axis=1, inplace=True)
    x_test = test.values
    
    return x_train, y_train, x_test, y_test
        
    
```
